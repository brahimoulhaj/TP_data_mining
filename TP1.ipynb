```python
#__Question2__#
X, y = datasets.make_blobs(n_samples=1000, centers=4, n_features=2)
```

```python
#__Question3__#
fig = plt.figure()
plt.scatter(X[:,0], X[:,1], c=y)
plt.colorbar()
plt.title('le titre de la figure')
plt.xlabel('X')
plt.ylabel('y')
plt.xlim(-15,15)
plt.ylim(-15,15)
plt.show()
```

```python
#__Question4__#
X_100, y_100 = datasets.make_blobs(n_samples=100, centers=2, n_features=2)
X_500, y_500 = datasets.make_blobs(n_samples=500, centers=3, n_features=2)
```

```python
X_600 = np.vstack((X_100, X_500))
y_600 = np.hstack((y_100, y_500))
```

```python
def my_scatter(X, y, title):
    plt.scatter(X[:,0], X[:,1], c=y)
    plt.colorbar()
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('y')
    plt.xlim(-15,15)
    plt.ylim(-15,15)
    plt.show()
```

```python
my_scatter(X_100, y_100, 'les 100 donnees')
my_scatter(X_500, y_500, 'les 500 donnees')
my_scatter(X_600, y_600, 'la concaténation des deux')
```

```python
import numpy as np
from sklearn import *
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [1, -1, 2],
    [2, 0, 0],
    [0, 1, -1]
])
print(X)

m = X.mean() #moyenne
print(f"\nmoyenne = {m}")

v = X.std() #variance
print(f"variance = {v}\n")

X_normalized = preprocessing.scale(X) #normaliser X
print(X_normalized)
# les donnees devient dans l'interval [-2,2]

m_normalized = X_normalized.mean()
v_normalized = X_normalized.std()
print(f"\nmoyenne = {m_normalized} ~= {round(m_normalized, 2)}\nvariance = {v_normalized}")
```

```python
X2 = np.array([[1, -1, 2], [2, 0, 0], [0, 1, -1]])
print(X2)

m2 = X2.mean(0)
print(f"\nmoyenne sur les variables = {m2}\n")

X2_normalized = preprocessing.minmax_scale(X2, feature_range=(0,1))
print(X2_normalized)

print(f"\nmoyenne sur les variables = {X2_normalized.mean(0)}")
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target

fig = plt.figure(figsize=(16,8))
plt.subplot(231); plt.scatter(X[:,0], X[:,1], c=y); plt.colorbar(); plt.xlabel('X'); plt.ylabel('y')
plt.subplot(232); plt.scatter(X[:,0], X[:,2], c=y); plt.colorbar(); plt.xlabel('X'); plt.ylabel('y')
plt.subplot(233); plt.scatter(X[:,0], X[:,3], c=y); plt.colorbar(); plt.xlabel('X'); plt.ylabel('y')
plt.subplot(234); plt.scatter(X[:,1], X[:,2], c=y); plt.colorbar(); plt.xlabel('X'); plt.ylabel('y')
plt.subplot(235); plt.scatter(X[:,1], X[:,3], c=y); plt.colorbar(); plt.xlabel('X'); plt.ylabel('y')
plt.subplot(236); plt.scatter(X[:,2], X[:,3], c=y); plt.colorbar(); plt.xlabel('X'); plt.ylabel('y')
plt.show()

#la meilleure visualisation est la combinaison (2,3) car les donnees sont separées plus que les autres
```

```python
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

#help(LDA)
#help(PCA)

pca = PCA(n_components=2)
lda = LDA(n_components=2)

irisPCA = pca.fit(X).transform(X)
irisLDA = lda.fit(X,y).transform(X)

fig = plt.figure(figsize=(16,8))
plt.subplot(121); plt.scatter(irisPCA[:,0], irisPCA[:,1], c=y); plt.title('P C A'); plt.xlabel('X'); plt.ylabel('y')
plt.subplot(122); plt.scatter(irisLDA[:,0], irisLDA[:,1], c=y); plt.title('L D A'); plt.xlabel('X'); plt.ylabel('y')
plt.show()

# les deux visualisations sont presque les mêmes
```
